```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

G.add_edges_from([('1', '2'), ('1', '3'), ('1', '5'), ('1', '6'), ('1', '7'), ('1', '10'),
                  ('2', '3'), ('2', '4'),('2', '8'),
                  ('2', '9'), ('2', '10'),('3', '8'),
                  ('3', '10'), ('3', '2'),('3', '9'),
                  ('4', '2'), ('4', '10'),('4', '7'),
                  ('4', '8'), ('5', '1'),('5', '6'),
                  ('5', '9'), ('5', '10'),
                  ('6', '1'), ('6', '5'),('6', '8'),
                  ('6', '9'), ('7', '9'),('7', '10'),('8', '10')])

#pos=nx.circular_layout(G)
G2 = G.to_undirected()
plt.figure(figsize =(10, 10))
#pos=nx.circular_layout(G2)
nx.draw_networkx(G2, with_labels = True)
```

```python
#Degree centrality
d=nx.degree_centrality(G2)
mean_arr = []
for i,k in enumerate(d):
  print(d[k])
  mean_arr.append(d[k])
```

```python
#Neighbour of node 2
# N = nx.neighbors(G2['2'])
# print(N)
print(list(G.neighbors('2')))
```

```python
#average degree of graph
import statistics
x = statistics.mean(mean_arr)
print("Mean is :", x)
```

```python
#density of graph
density = nx.density(G2)
print("Network density:", density)
nx.draw_networkx(G2, with_labels=True)
```

```python
#closness of 10
x = nx.closeness_centrality(G2,'10')
print(x)
```

```python
#Find all the paths to reach 4 from 6
print('path to reach from node 4 to 6 : ')
paths = nx.all_simple_paths(G2, source='4', target='6')
print(list(paths))
print()

```

```python
#	Find the longest shortest path between any two nodes
fell_whitehead_path = nx.shortest_path(G2, source="3", target="7")

print("Shortest path:", fell_whitehead_path)
print("Length of that path:", len(fell_whitehead_path)-1)
```

```python
#	Find the Betweenness centrality of Node 1.
betweenness_centrality = nx.betweenness_centrality(G2)
betweenness_centrality

print(betweenness_centrality['1'])
```

```python
#	Find the person who has maximum number of connections (friends)
degree_dict = dict(G2.degree(G2.nodes()))
nx.set_node_attributes(G2, degree_dict, 'degree')

sorted_degree = sorted(degree_dict.items(), reverse=True)
print("Top nodes by degree:")
for d in sorted_degree[:20]:
    print(d)
```
